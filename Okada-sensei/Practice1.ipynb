```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
%matplotlib inline
```

```python
temp=np.array([55, 60, 65, 70, 75, 80, 85, 90])
rate=np.array([45,80,92,114,141,174,202,226])

slope, intercept, r_val, p_val, std_err = linregress(temp, rate)
```

```python
print(f'The equation of regression line is y={slope:.3f}x+{intercept:.3f}.')
```

```python
X_plot = np.linspace(50,100,100)
Y_plot = slope*X_plot+intercept
plt.plot(X_plot, Y_plot, color='r')
plt.scatter(temp,rate,label='rate')
plt.title("temp vs rate")
plt.xlabel('temp')
plt.ylabel('rate')
plt.legend()
plt.show()
```
